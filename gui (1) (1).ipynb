```python
import ipywidgets as widgets
from IPython.display import display, HTML

class VectorGraphicsEditor:
    def __init__(self):
        self.canvas = widgets.Output()
        self.selected_shape = None
        self.shapes = []

        self.create_tools()

    def create_tools(self):
        self.shape_dropdown = widgets.Dropdown(
            options=["Circle", "Square", "Triangle"],
            value="Circle",
            description="Shape:"
        )

        self.color_picker = widgets.ColorPicker(
            concise=True,
            description='Color:',
            value='black'
        )

        self.draw_button = widgets.Button(description="Draw")
        self.draw_button.on_click(self.draw_shape)

        self.delete_button = widgets.Button(description="Delete")
        self.delete_button.on_click(self.delete_shape)

        self.tools_layout = widgets.VBox([
            self.shape_dropdown,
            self.color_picker,
            self.draw_button,
            self.delete_button
        ])

        display(widgets.HBox([self.tools_layout, self.canvas]))

    def draw_shape(self, button):
        shape = self.shape_dropdown.value
        color = self.color_picker.value
        self.shapes.append((shape, color))
        with self.canvas:
            self.canvas.clear_output()
            self._draw_shapes()

    def delete_shape(self, button):
        self.shapes = []
        with self.canvas:
            self.canvas.clear_output()

    def _draw_shapes(self):
        html_str = """
        <canvas id="myCanvas" width="400" height="400" style="border:1px solid #000000;">
        Your browser does not support the HTML5 canvas tag.
        </canvas>
        <script>
        var c = document.getElementById("myCanvas");
        var ctx = c.getContext("2d");
        """
        for shape, color in self.shapes:
            if shape == "Circle":
                html_str += f"""
                ctx.beginPath();
                ctx.arc(200, 200, 50, 0, 2 * Math.PI);
                ctx.fillStyle = "{color}";
                ctx.fill();
                """
            elif shape == "Square":
                html_str += f"""
                ctx.fillStyle = "{color}";
                ctx.fillRect(150, 150, 100, 100);
                """
            elif shape == "Triangle":
                html_str += f"""
                ctx.beginPath();
                ctx.moveTo(300,200);
                ctx.lineTo(350,150);
                ctx.lineTo(250,150);
                ctx.fillStyle = "{color}";
                ctx.fill();
                """
        html_str += "</script>"
        display(HTML(html_str))

editor = VectorGraphicsEditor()
```
